# Supplementary: Plotting the Language Generation Results

Running this exercise successfully depednds on having the following files available.
* `lstm.csv` (dropout = 0.0) and `lstm_dropout.csv` (dropout > 0.0) - Obtained by running `lstm_extensions.ipynb` with the LSTM algorithm (run only the standard LSTM when you have to choose between standard LSTM, LSTM with peepholes and GRUs) 
* `lstm_peephole_dropout.csv` (dropout > 0.0) - Obtained by running `lstm_extensions.ipynb` with the LSTM with peepholes algorithm
* `gru_dropout.csv` (dropout > 0.0) - Obtained by running `lstm_extensions.ipynb` with the GRU algorithm
* `lstm_word2vec_beamsearch_dropout.csv` (dropout > 0.0) - Obtained by running the code in `lstm_word2vec.ipynb`
* `lstm_beam_search_word2vec_rnn_api.csv` - Obtained by running the code in `lstm_word2vec_rnn_api.ipynb`


```python
# These are all the modules we'll be using later. Make sure you can import them
# before proceeding further.
%matplotlib inline
import numpy as np
from matplotlib import pylab
import csv
```

```python
# Loading the files

# LSTM no dropout
with open('lstm.csv', 'rt') as f:
    reader = csv.reader(f,delimiter=',')
    for r_i,row in enumerate(reader):
        if r_i == 0:
            train_lstm_no_dropout_perplexity = [float(s) for s in row]
        if r_i == 2:
            valid_lstm_no_dropout_perplexity = [float(s) for s in row]

"""
# LSTM with dropout
with open('lstm_dropout.csv', 'rt') as f:
    reader = csv.reader(f,delimiter=',')
    for r_i,row in enumerate(reader):
        if r_i == 0:
            train_lstm_perplexity = [float(s) for s in row]
        if r_i == 2:
            valid_lstm_perplexity = [float(s) for s in row]

"""

# LSTM with peephole + Dropout
with open('lstm_peephole_dropout.csv', 'rt') as f:
    reader = csv.reader(f,delimiter=',')
    for r_i,row in enumerate(reader):
        if r_i == 0:
            train_lstm_peephole_perplexity = [float(s) for s in row]
        if r_i == 2:
            valid_lstm_peephole_perplexity = [float(s) for s in row]


# GRU with dropout
with open('gru_dropout.csv', 'rt') as f:
    reader = csv.reader(f,delimiter=',')
    for r_i,row in enumerate(reader):
        if r_i == 0:
            train_gru_perplexity = [float(s) for s in row]
        if r_i == 2:
            valid_gru_perplexity = [float(s) for s in row]

         
# LSTM + Word2vec + Dropout
with open('lstm_word2vec_beamsearch_dropout.csv', 'rt') as f:
    reader = csv.reader(f,delimiter=',')
    for r_i,row in enumerate(reader):
        if r_i == 0:
            train_lstm_word2vec_perplexity = [float(s) for s in row]    
        if r_i == 2:
            valid_lstm_word2vec_perplexity = [float(s) for s in row]
    """  
            
# Deep LSTM + Word2vec + Dropout
with open('lstm_beam_search_word2vec_rnn_api.csv', 'rt') as f:
    reader = csv.reader(f,delimiter=',')
    for r_i,row in enumerate(reader):
        if r_i == 0:
            train_lstm_word2vec_deep_perplexity = [float(s) for s in row]  
        if r_i == 2:
            valid_lstm_word2vec_deep_perplexity = [float(s) for s in row]
     """
```

```python
# Plots without the Word2vec related results
pylab.figure(figsize=(15,10)) 

# Plot the training perplexities
pylab.subplot(211)
"""
pylab.plot(np.arange(len(train_lstm_perplexity)), 
           np.log(train_lstm_perplexity), marker='o', 
           label = "LSTM + Dropout")
"""

pylab.plot(np.arange(len(train_lstm_no_dropout_perplexity)), 
           np.log(train_lstm_no_dropout_perplexity), linestyle='--', 
           marker='>', label = "LSTM")


pylab.plot(np.arange(len(train_lstm_peephole_perplexity)), 
           np.log(train_lstm_peephole_perplexity), marker='+', 
           markersize=10, label='LSTM (Peephole) + Dropout')


pylab.plot(np.arange(len(train_gru_perplexity)), 
           np.log(train_gru_perplexity), marker='x', 
           linestyle='-.', label='GRU + Dropout')

pylab.legend(loc=1,fontsize=16)
pylab.title('Train Perplexity Behavior for LSTM Variants',fontsize=18)
pylab.ylabel('Log Perplexity',fontsize=18)

# Plot the validation perplexities
pylab.subplot(212)
"""
pylab.plot(np.arange(len(valid_lstm_perplexity)), 
           np.log(valid_lstm_perplexity), marker='o', 
           label = "LSTM + Dropout")
"""

pylab.plot(np.arange(len(valid_lstm_no_dropout_perplexity)), 
           np.log(valid_lstm_no_dropout_perplexity), linestyle='--', 
           marker='>', label = "LSTM")

pylab.plot(np.arange(len(valid_lstm_peephole_perplexity)), 
           np.log(valid_lstm_peephole_perplexity), marker='+', 
           markersize=10, label='LSTM (Peephole) + Dropout')


pylab.plot(np.arange(len(valid_gru_perplexity)), 
           np.log(valid_gru_perplexity), marker='x', 
           linestyle='-.', label='GRU + Dropout')

pylab.legend(loc=1,fontsize=16)
pylab.title('Valid Perplexity Behavior for LSTM Variants',fontsize=18)
pylab.ylabel('Log Perplexity',fontsize=18)
pylab.xlabel('Time Step',fontsize=18)
pylab.savefig('lstm_perplexity_all.png')
pylab.show()
```
